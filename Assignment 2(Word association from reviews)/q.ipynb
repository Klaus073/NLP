```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
import numpy as np
import pandas as pd
```

```python
data = open('reviews.txt').read()
```

```python
data=data.split('\n') 
```

```python
del data[0]
del data[-1]
```

```python
for x in range(len(data)):
    data[x] = data[x].replace('"','')
    data[x] = data[x].strip()
```

```python
vec = CountVectorizer(max_features = 15, stop_words='english')
output = vec.fit_transform(data)
matrix = output.toarray()
```

```python
datavec = pd.DataFrame(matrix)
datavec
```

```python
X = []
for i in range(len(datavec.iloc[0])):
    #print((datavec.iloc[0]))
    for j in range(i+1, len(datavec.iloc[0])):
        temp = (i,j)
        temp2 = metrics.mutual_info_score(datavec[i],datavec[j])
        temp3 = (temp,temp2)
        print(temp3)
        X.append(temp3)
```

```python
Scores = []
for i in X:
    
    Scores.append(i[1])
```

```python
Scores
```

```python
Y = np.argsort(Scores, axis = 0)
# print(Scores[Y[-1]])
TopScores = []
```

```python
for x in range(1,len(X)):
    num = -x
    TopScores.append(X[Y[num]])
```

```python
TopScores
```

```python
print("Top 50 word pair associations are: ")
for x in range(50):
    print("Pair",x+1,end=": ")
    # print(vec.get_feature_names()[X[0][0][0])
    print(vec.get_feature_names()[TopScores[x][0][0]], end = " ")
    print(vec.get_feature_names()[TopScores[x][0][1]], end = "\n")
```
